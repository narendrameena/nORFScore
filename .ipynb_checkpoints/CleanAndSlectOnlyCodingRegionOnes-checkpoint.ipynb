```python
import numpy as np
import pandas as pd
import dask.dataframe as dd
```

```python
#classification of nORFs
classified = pd.read_csv("../data/input/nORF/nORFclassification.tsv",sep="\t")
classified.head()
```

```python
len(classified)
```

```python
classified = classified[classified['codingRegion'] == True]
classified.head()
```

```python
len(classified)
```

```python
#mapped variants to nORFs
varinats_hgmd = pd.read_csv("../analysis/variantsMappedOnnORFs/hgmd/hgmd_mapped_nORF.txt",sep="\t",header=None, converters={i: str for i in range(250)})
varinats_hgmd = varinats_hgmd[[0,1,3,4,11]]
varinats_hgmd.columns = ['Chr', 'End','Ref','Alt','novelORF_ID']
varinats_hgmd.head()
```

```python
varinats_clinvar = pd.read_csv("../analysis/variantsMappedOnnORFs/clinvar/clinvar_mapped_nORF.txt",sep="\t",header=None, converters={i: str for i in range(250)})
varinats_clinvar = varinats_clinvar[[0,1,3,4,11]]
varinats_clinvar.columns = ['Chr', 'End','Ref','Alt','novelORF_ID']
varinats_clinvar.head() 
```

```python
varinats_human_drive = pd.read_csv("../analysis/variantsMappedOnnORFs/humanDrive/humanDrive_mapped_nORF.txt",sep="\t",header=None, converters={i: str for i in range(250)})
varinats_human_drive = varinats_human_drive[[0,1,3,4,8]]
varinats_human_drive.columns = ['Chr', 'End','Ref','Alt','novelORF_ID']
varinats_human_drive.head() 


```

```python
#hgmd variants from coding region 

coding_variants_hgmd = pd.merge(classified, varinats_hgmd, how='inner',on=['novelORF_ID'])
coding_variants_hgmd = coding_variants_hgmd[['Chr', 'End','Ref','Alt','novelORF_ID']] 
coding_variants_hgmd.head()
```

```python
#clinvar variants from coding regions 

coding_variants_clinvar = pd.merge(classified, varinats_clinvar, how='inner',on=['novelORF_ID'])
coding_variants_clinvar = coding_variants_clinvar[['Chr', 'End','Ref','Alt','novelORF_ID']] 
coding_variants_clinvar.head()
```

```python
#human drive from coding region and drop duplicates 
coding_variants_human_drive = pd.merge(classified, varinats_human_drive, how='inner',on=['novelORF_ID'])
coding_variants_human_drive = coding_variants_human_drive[['Chr', 'End','Ref','Alt','novelORF_ID']] 
coding_variants_human_drive = coding_variants_human_drive[['Chr', 'End','Ref','Alt']].drop_duplicates()
coding_variants_human_drive.head()
```

```python
#concate two dataframe and drop duplicates 
coding_variants_hgmd_clinvar = pd.concat([coding_variants_hgmd,coding_variants_clinvar],axis=0, join='outer',ignore_index=True)
coding_variants_hgmd_clinvar = coding_variants_hgmd_clinvar[['Chr', 'End','Ref','Alt']].drop_duplicates() 
coding_variants_hgmd_clinvar.head()



```

```python
#attributes
pathogenic = pd.read_csv("../analysis/trainingDataSet/pathogenic_with_all_attributes_cat_uniqe.tsv",sep="\t", 
                         converters={i: str for i in range(250)})
pathogenic.head()
```

```python
pathogenic = pathogenic.loc[pathogenic['Ref'].isin(['A','T','G','C'])]
```

```python
pathogenic = pathogenic.loc[pathogenic['Alt'].isin(['A','T','G','C'])]
```

```python
tbfs = dd.read_csv("../data/input/tbfs/snp2tfbs_JASPAR_CORE_2014_vertwithout_chr.bed",sep="\t",header=None,
                   names=["Chr","Start","End","Ref","Alt","rsid","tfbs","tfbsName","tbfsScore"], converters={i: str for i in range(250)})
```

```python
tbfs.head()
```

```python
pathogenic[["Chr","End","Ref","Alt"]].head()
```

```python
pathogenic = dd.merge(pathogenic,tbfs, how='left', on=['Chr', 'End', 'Ref','Alt']).compute()
```

```python
coding_pathogenic = dd.merge(pathogenic, coding_variants_hgmd_clinvar,how='inner', on=['Chr', 'End', 'Ref','Alt']) 
coding_pathogenic.head()
```

```python
list(coding_pathogenic.columns)
```

```python
case = case.replace('null', np.nan, regex=True)
case = case.replace('Null', np.nan, regex=True)
case = case.replace('NA', np.nan, regex=True)
case = case.replace('.', np.nan)
case = case.replace('-', np.nan)
case = case.replace('*', np.nan)

case.to_csv("../analysis/trainingDataSet/slected_pathogenic_Coding_region_only.csv", sep="\t",index=False)

case.head()
```
